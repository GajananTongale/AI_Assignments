```python
import numpy as np

def remove_outliers_iqr(df, columns):
    for col in columns:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR
        df = df[(df[col] >= lower) & (df[col] <= upper)]
    return df



selected_columns = ['Daily_Usage_Hours', 'Weekend_Usage_Hours','Exercise_Hours','Screen_Time_Before_Bed']

df = remove_outliers_iqr(df, selected_columns)

```
